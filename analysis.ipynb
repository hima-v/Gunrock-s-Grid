# Distance vs. Price Analysis

This notebook analyzes the relationship between various distance metrics and housing prices in the UC Davis area using scatter plots with trendlines.

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
df = pd.read_csv('./enriched_listings.csv')

print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst few rows:")
df.head()
```

## Distance vs. Price - Memorial Union

```python
data_mu = df[['dist_to_memorial_union_mu', 'price_per_bed']].dropna()

# Calculate trendline
z = np.polyfit(data_mu['dist_to_memorial_union_mu'], data_mu['price_per_bed'], 1)
p = np.poly1d(z)
x_trend = np.linspace(data_mu['dist_to_memorial_union_mu'].min(), data_mu['dist_to_memorial_union_mu'].max(), 100)
y_trend = p(x_trend)

# Calculate correlation
correlation = data_mu['dist_to_memorial_union_mu'].corr(data_mu['price_per_bed'])

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=data_mu['dist_to_memorial_union_mu'],
    y=data_mu['price_per_bed'],
    mode='markers',
    name='Listings',
    marker=dict(size=8, color='blue', opacity=0.6)
))
fig.add_trace(go.Scatter(
    x=x_trend,
    y=y_trend,
    mode='lines',
    name='Trendline',
    line=dict(color='red', width=2)
))

fig.update_layout(
    title=f'Distance to Memorial Union vs Price per Bed (Correlation: {correlation:.3f})',
    xaxis_title='Distance to Memorial Union (miles)',
    yaxis_title='Price per Bed ($)',
    template='plotly_white',
    height=500
)
fig.show()

print(f"Correlation: {correlation:.3f}")
print(f"Trendline equation: y = {z[0]:.2f}x + {z[1]:.2f}")
```

```python
import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("enriched_listings.csv")

# Keep rows with valid values
bubble_df = df.dropna(subset=["sqft", "price_total", "bedrooms"]).copy()
bubble_df = bubble_df[
    (bubble_df["sqft"] > 0) &
    (bubble_df["price_total"] > 0) &
    (bubble_df["bedrooms"] > 0)
]

# Bubble chart
fig = px.scatter(
    bubble_df,
    x="sqft",
    y="price_total",
    size="bedrooms",
    color="bedrooms",
    hover_name="address",
    hover_data=["complex_name", "baths", "price_per_bed", "price_per_sqft"],
    title="Size vs Cost Efficiency (All Listings)",
    labels={
        "sqft": "Square Feet",
        "price_total": "Total Rent ($)",
        "bedrooms": "Bedrooms"
    },
    size_max=40
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Square Feet",
    yaxis_title="Total Rent ($)",
    legend_title="Bedrooms"
)

fig.show()
```
