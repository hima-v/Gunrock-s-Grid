```python
import pandas as pd
import plotly.express as px

# Create a copy so original data is untouched
df2 = df.copy()

# Merge categories
df2["laundry_type"] = df2["laundry_type"].replace({
    "on-site": "shared/on-site",
    "shared": "shared/on-site"
})

# Calculate averages
laundry_avg = (
    df2.groupby("laundry_type")["price_per_bed"]
    .mean()
    .reset_index()
)

# Plot
fig = px.bar(
    laundry_avg,
    x="laundry_type",
    y="price_per_bed",
    title="Average Rent per Bedroom by Laundry Type"
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Example counts
neighborhood_counts = df["neighborhood"].value_counts().reset_index()
neighborhood_counts.columns = ["Neighborhood", "Count"]

fig = px.pie(
    neighborhood_counts,
    names="Neighborhood",
    values="Count",
    hole=0.45   # makes donut chart
)

# Center title
fig.update_layout(
    annotations=[dict(
        text="Neighborhood<br>Distribution",
        x=0.5,
        y=0.5,
        font_size=18,
        showarrow=False
    )],

    # Move legend closer
    legend=dict(
        x=0.7,
        y=0.7
    ),

    title="Distribution of Listings by Neighborhood"
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

#So apartments get a higher score if they have:
#  More space per student
# Lower rent per student
# Closer distance to campus

df = pd.read_csv("/Users/hetvi/Documents/Stats/Gunrock-s-Grid/enriched_listings.csv")

# Remove invalid rows
plot_df = df.dropna(subset=["sqft","bedrooms","price_per_bed","dist_to_mu","complex_name"])

# Avoid division errors
plot_df = plot_df[(plot_df["bedrooms"] > 0) & (plot_df["dist_to_mu"] > 0)]

# Calculate space per bedroom
plot_df["sqft_per_bed"] = plot_df["sqft"] / plot_df["bedrooms"]

# Student value score
plot_df["student_score"] = plot_df["sqft_per_bed"] / (
    plot_df["price_per_bed"] * plot_df["dist_to_mu"]
)

# Top apartments
top = plot_df.sort_values("student_score", ascending=False).head(10)

fig = px.bar(
    top,
    x="student_score",
    y="complex_name",
    orientation="h",
    color_discrete_sequence=["#582a9d"],
    title="Top 10 Best Student Value Apartments",
    labels={
        "student_score": "Student Value Score",
        "complex_name": "Apartment"
    },
    hover_data=["sqft_per_bed","price_per_bed","dist_to_mu"]
)

fig.update_layout(yaxis={"categoryorder":"total ascending"})

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# So apartments get a higher score if they have:
# More space per student
# Lower rent per student
# Closer distance to campus

df = pd.read_csv("/Users/hetvi/Documents/Stats/Gunrock-s-Grid/enriched_listings.csv")

# Remove invalid rows
plot_df = df.dropna(subset=["sqft","bedrooms","price_per_bed","dist_to_mu","address"])

# Avoid division errors
plot_df = plot_df[(plot_df["bedrooms"] > 0) & (plot_df["dist_to_mu"] > 0)]

# Calculate space per bedroom
plot_df["sqft_per_bed"] = plot_df["sqft"] / plot_df["bedrooms"]

# Student value score
plot_df["student_score"] = plot_df["sqft_per_bed"] / (
    plot_df["price_per_bed"] * plot_df["dist_to_mu"]
)

# Top apartments
top = plot_df.sort_values("student_score", ascending=False).head(10)

# Print row numbers from the original dataset
print("Row numbers of top apartments:")
print(top.index.tolist())
fig = px.bar(
    top,
    x="student_score",
    y="address",
    orientation="h",
    color_discrete_sequence=["#582a9d"],
    title="Top 10 Best Student Value Apartments",
    labels={
        "student_score": "Student Value Score",
        "address": "Apartment Address"
    },
    hover_data=["sqft_per_bed","price_per_bed","dist_to_mu"]
)

fig.update_layout(yaxis={"categoryorder":"total ascending"})

fig.show()
```
